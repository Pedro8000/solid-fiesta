## Praticando

```python
df.shape
```

```python
select_max = df.loc[df['price'].idxmax()]  ## Carro mais caro
select_max
```

```python
select_min = df.loc[df['price'].idxmin()] ## Carro mais barato
select_min
```

```python
toyota = df.loc[(df['brand'] == 'toyota')] ## Localizando todos os carros da toyota
```

```python
toyota.drop(['price', 'vin', 'lot', 'state', 'country', 'condition', 'mileage', 'color','title_status'], axis=1)
```

```python
ford = df.loc[(df['brand'] == 'ford')]  ## Localizando todos os carros da ford
```

```python
ford.drop(['price', 'vin', 'lot', 'state', 'country', 'condition', 'mileage', 'color','title_status'], axis=1)
```

```python
condicao = df.groupby(['title_status']).count()  ## Mostrando quantos veículos existem em  cada uma das duas condições
```

```python
condicao.drop(['price','model','brand', 'year', 'vin', 'lot', 'state', 'country', 'condition', 'mileage', 'color'], axis=1)
```

```python
cores = df.groupby(['color']).count()  ## Mostrando quantos carros de cada cor estão no dataframe
```

```python
cores.drop(['price','model','brand', 'title_status', 'year', 'vin', 'lot', 'state', 'country', 'condition', 'mileage'],  axis=1)
```
